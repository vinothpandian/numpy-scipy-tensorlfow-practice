```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
a = tf.constant(2)
b = tf.constant(3)
x = tf.add(a, b)

with tf.Session() as sess:
    print(sess.run(x))
```

```python
a = tf.constant(2, name="a")
b = tf.constant(3, name="b")
x = tf.add(a, b, name="add")

with tf.Session() as sess:
    writer = tf.summary.FileWriter('./graphs', sess.graph)
    print(sess.run(x))
writer.close()
```

```python
a = tf.constant([2,2], name="a")
b = tf.constant([[0, 1],[2, 3]], name="b")
x = tf.multiply(a, b, name="multiply")

with tf.Session() as sess:
    print(sess.run(x))
```

```python
zeros = tf.zeros([2, 3], tf.int32, name="zeros")

with tf.Session() as sess:
    print(sess.run(zeros))
```

```python
input_tensor = np.arange(9).reshape(3,3)
print(input_tensor)

zeros = tf.zeros_like(input_tensor, name="zeros_like")

with tf.Session() as sess:
    print(sess.run(zeros))
```

```python
input_tensor = np.arange(9).reshape(3,3)
print(input_tensor)

ones = tf.ones([2, 2], name="ones")
ones_like = tf.ones_like(input_tensor, name="ones_like")

with tf.Session() as sess:
    print(sess.run(ones))
    print(sess.run(ones_like))
```

```python
filled = tf.fill([2, 2], 8., name="eights")

with tf.Session() as sess:
    print(sess.run(filled))
```

```python
linear_space = tf.lin_space(10., 11., 5, name="linear_space")
evens = tf.range(0, 10, 2, name="evens")
ranged = tf.range(5, name="range")

with tf.Session() as sess:
    print(sess.run(linear_space))
    print(sess.run(evens))
    print(sess.run(ranged))
```

```python
rand_norm = tf.random_normal([1, 100], mean=0., stddev=3., name="rand_norm")

with tf.Session() as sess:
    rand_norm = sess.run(rand_norm)
    
print(rand_norm)
```

```python
x = tf.constant(np.arange(1000), name="x")

with tf.Session() as sess:
    print(sess.graph.as_graph_def())
```

```python
g = tf.Graph()

with g.as_default():
    x = tf.get_variable("x_var", initializer=tf.constant(np.arange(1000)))

with tf.Session(graph=g) as sess:
    print(sess.graph.as_graph_def())
```

```python
W = tf.Variable(10)
double_W = W.assign(2 * W)
triple_W = W.assign(3 * W)

with tf.Session() as sess:
    sess.run(W.initializer)
    sess.run(double_W)
    print(sess.run(W))
    sess.run(double_W)
    print(sess.run(W))
    sess.run(triple_W)
    print(W.eval())

```

```python
a = tf.placeholder(tf.float32, shape=[3], name="a")
b = tf.Variable(tf.fill([3], 5.), name="b")

c = tf.add(a,b)

with tf.Session() as sess:
    sess.run(b.initializer)
    c_out = sess.run(c, feed_dict={ a: np.ones(3)})
    
    
print(c_out)

```
